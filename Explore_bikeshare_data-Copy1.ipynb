```r
ny = read.csv('new_york_city.csv')
wash = read.csv('washington.csv')
chi = read.csv('chicago.csv')
```

```r
# Using rbind to create a new data set "mydf" to merge the New York and Chicago dataset.
mydf <- rbind(ny,chi)
table(mydf$Gender)
```

```r
#Using a histogram to chart the gender distribution.
library("ggplot2")
qplot(data=mydf,x=Gender,fill=I("orange"), main="Gender distribution of users in New York and Chicago", ylab="Frequency")
```

1. It is evident from data of both cities, New York and Chicago that males are the largest adopters of Bike Share services. 

2. There are entries in the data where the gender of the user has not been captured, and hence there are 3 datasets in the histogram.

Question 2

Is there a relationship between Gender and Trip Duration? 

```r
qplot(data=mydf,x=Gender, y=Trip.Duration/60, colour=Gender, size=I(1),ylim=c(0,3000), ylab= "Trip Duration", main="Co-relation between Gender and Trip Duration")
```

1. Users who have not shared their gender information, indicated in red, have the longest trip duration. 
2. Between males and females, males are using Bike Share services to travel a longer distance. 
3. I have converted trip duration from seconds to minutes to make it easier to visualise the data. 

### Question 3

**Is there a corelation between the gender and age of the user?**

```r
qplot(data=mydf,x=Gender,y= Birth.Year, geom="boxplot", ylim=c(1950,2000), ylab="Birth Year", main= "Co-relation between Gender and Birth Year")
```

1. The median Birth Year for females is 1983, whereas the median Birth Year for Males is 1981 and for the missing Birth Year is 1980.
2. The Inter Quartile Range in Years is 1972 to 1989 for Females and 1970 to 1988 for Males. This indicates that the middle 50% of the values for both Males and females are fairly similar.
3. In terms of Age, both Males and Females have shown a similar trend of adoption of Bike Share Services.
